```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the sigmoid function
x = np.linspace(-10, 10, 400)
y = 1 / (1 + np.exp(-x))

# Disable LaTeX rendering by setting usetex to False
plt.rc('text', usetex=False)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='black', lw=2)
plt.axhline(y=0.5, color='blue', linestyle='--', label='Threshold at 0.5')

# Add labels and title
plt.title('Sigmoid Function with Threshold at 0.5', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Show legend and grid
plt.legend(fontsize=12)
plt.grid(True)

# Display the plot
plt.show()
```
